```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
import csv
```

```python
# Create an URL object
url = 'https://www.mapcrow.info/united_states.html'
# Create object page
page = requests.get(url)
```

```python
page
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
# Obtain information from tag <table>
table1 = soup.find('table', border="1")
table1
```

```python
headers = []
for i in table1.find_all('td'):
    table = i.text
    headers.append(table)
```

```python
headers
```

```python
df = pd.DataFrame(index=np.arange(32), columns=np.arange(32))
x=[0]*32
dummy=0
for i in range(32):
    for j in range(32):
        x[j]=headers[dummy]
        df[i] = x
        dummy=dummy+1
```

```python
x=df.set_index(0).T.to_dict('index')
```
